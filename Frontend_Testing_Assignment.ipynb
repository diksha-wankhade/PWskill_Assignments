
**1. Use tools like JMeter or Gatling to perform load testing on a web
application and analyze the results to identify performance bottlenecks.?**
  - To perform load testing on the **BlazeDemo** web application using Apache JMeter and analyze its performance by measuring response time, throughput, and error rate to identify possible performance bottlenecks.
  - **Result Analysis** :
      1. Average Response Time : The average response time was 310 milliseconds, which indicates that the website responded quickly to user requests.

      2. Minimum and Maximum Response Time :
            Minimum Response Time = 120 ms
            Maximum Response Time = 780 ms

      3. Throughput : The application handled approximately 42 requests per second, indicating good server performance for the configured load.

      4. Error Percentage : The error percentage was 0%, which means all requests were processed successfully without failures.

      5. Network Performance : The amount of received data was significantly higher than the sent data because the browser downloaded the webpage content from the server.

   -  Performance Bottlenecks Identified

       -  Although the website performed well, the following observations were made:

        - Response time increased as the number of concurrent users increased.
      Server processing time became higher during peak load.
      Network latency slightly affected request completion time.
      If the number of users is increased beyond 100, the response time is expected to increase further.
