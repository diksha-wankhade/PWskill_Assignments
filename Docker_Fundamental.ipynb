**1. Use Docker Compose to define and run multi-container applications?**
  - Docker Compose is a tool that allows developers to define and manage multi-container Docker applications using a single YAML configuration file (docker-compose.yml).
  - Instead of starting each container individually, Docker Compose enables all required services to be launched with a single command. It simplifies application deployment, networking, and dependency management.
  - Prerequisites :
      - Docker Desktop installed
      - Docker Compose installed (included with Docker Desktop)
      - Basic knowledge of Docker containers
  - Deploy a simple application with two containers:
     ```
       Nginx (Web Server)
       MySQL (Database Server)

  - Docker Compose will automatically:
      - Create a network for communication between containers.
      - Start both containers.
      - Manage their lifecycle.
  - Step 1: Create an HTML File: Create a file named index.html.
      ```
        <!DOCTYPE html>
        <html>
        <head>
            <title>Docker Compose Demo</title>
        </head>
        <body>
            <h1>Welcome to Docker Compose!</h1>
            <p>Multi-container application is running successfully.</p>
        </body>
        </html>
   - Step 2: Create the docker-compose.yml File
        ```
          version: "3.9"

          services:

            web:
              image: nginx:latest
              container_name: nginx-server
              ports:
                - "8080:80"
              volumes:
                - ./index.html:/usr/share/nginx/html/index.html
              depends_on:
                - mysql

            mysql:
              image: mysql:8.0
              container_name: mysql-db
              restart: always

              environment:
                MYSQL_ROOT_PASSWORD: root123
                MYSQL_DATABASE: companydb
                MYSQL_USER: admin
                MYSQL_PASSWORD: admin123

              ports:
                - "3306:3306"

              volumes:
                - mysql-data:/var/lib/mysql

          volumes:
            mysql-data:
        ```    

  - Docker Compose Architecture
  ```
                Docker Compose
                       │
        ┌──────────────┴──────────────┐
        │                             │
   Nginx Container              MySQL Container
   (Web Server)                 (Database Server)
        │                             │
        └────────── Docker Network ───┘
  ```
- Advantages of Docker Compose :
  - Defines multi-container applications in a single YAML file.
  - Starts and stops all services with one command.
  - Automatically creates networks between containers.
  - Supports persistent storage using volumes.
  - Simplifies development, testing, and deployment.
  - Easy to scale services when required.
