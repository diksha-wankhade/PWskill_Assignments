**1. What are data structures, and why are they important?**
  - A data structure is a method of organizing and storing data so that it can be accessed and modified efficiently. Different data structures are designed for different types of operations, such as searching, sorting, inserting, or deleting data. Common data structures include arrays, lists, tuples, stacks, queues, trees, graphs, dictionaries, and sets.
  - Data structures are important because they improve the performance of programs by reducing the time and memory required to process data. They help developers write efficient, organized, and scalable applications. Choosing the right data structure can significantly improve the speed and effectiveness of a program.

**2. Explain the difference between mutable and immutable data types with examples**.
  - Mutable data types are objects whose values can be changed after they are created. For example, lists, dictionaries, and sets are mutable because elements can be added, removed, or modified.
  - Example:
     ```
      numbers = [10, 20, 30]
      numbers.append(40)
      print(numbers)

      Output:
      [10, 20, 30, 40]

  - Immutable data types cannot be modified once they are created. If a change is required, a new object is created instead. Examples include strings, tuples, integers, and floats.
  - Example:
    ```
      name = "Python"
      name = name + " Programming"
      print(name)

      Output:

      Python Programming
**3. What are the main differences between lists and tuples in Python?**
  - Lists and tuples are both used to store collections of items, but they have several differences.

  - A list is mutable, whereas a tuple is immutable.

  - Lists consume more memory because they allow modifications.
  - Tuples are faster than lists because they cannot be changed.
  - Lists are suitable for storing data that changes frequently, whereas tuples are ideal for fixed data.

      Example:
      ```
      my_list = [1, 2, 3]
      my_tuple = (1, 2, 3)
**4. Describe how dictionaries store data.**
    ```

     ```
**8. What is a hash table, and how does it relate to dictionaries in Python?**
 - A hash table is a data structure that stores data in the form of key-value pairs. It uses a hash function to calculate the storage location of each key, allowing data to be accessed very quickly.
 - Python dictionaries are implemented using hash tables. When a key is provided, Python calculates its hash value and directly locates the corresponding value instead of searching sequentially.
 - This makes dictionary operations such as searching, inserting, and deleting very efficient, with an average time complexity of O(1).
     ```

**14. How does the 'in' keyword work differently for lists and dictionaries?**
  - The in keyword is used to check whether an element exists in a collection, but it behaves differently for lists and dictionaries.
  - In a list, it checks whether the specified value is present among the list elements.
  - In a dictionary, it checks only whether the specified key exists, not the value.

    Example with List:
     ```
      numbers = [10, 20, 30]
      print(20 in numbers)

      Output:
      True
      ```
    Example with Dictionary:
    ```
      student = {

      }
      print("name" in student)

      Output:
      True
    ```

**17. Describe the time complexity of accessing elements in a dictionary.**
  - Python dictionaries use a hash table internally. Because of this, accessing an element using its key takes O(1) time on average, which means the lookup operation is very fast.
  - In rare situations where multiple keys produce the same hash value (known as collisions), the time complexity may become O(n). However, Python uses efficient collision-handling techniques, so this happens very rarely.
  - Therefore, dictionaries are considered one of the fastest data structures for searching data.

  - For example, storing marks of students in the order they were obtained is better suited for a list.

    ```
**19. Why are dictionaries considered unordered, and how does that affect data retrieval?**
  - Dictionaries were considered unordered because they stored data based on the hash values of keys rather than insertion order. This allowed very fast retrieval of values.
  - Starting with Python 3.7, dictionaries preserve the order in which items are inserted. However, values are still retrieved using their keys, not by their position or index.
  - This means users should access dictionary elements by specifying the key rather than relying on numerical indexes.

    Example:
    ```
    employee = {
        "ID": 101,
        "Name": "Raj",
        "Department": "IT"
    }

    print(employee["Department"])

  - In a dictionary, data is retrieved using a unique key instead of an index.
  - Lists require sequential indexing, whereas dictionaries provide direct access to values using keys, making lookups much faster for large datasets.
  - Example of a List:
    ```
      colors = ["Red", "Green", "Blue"]
      print(colors[1])

      Output:
      Green

  - Example of a Dictionary:
    ```
      student = {
          "Name": "Anita",
          "Age": 20
      }

      print(student["Age"])
