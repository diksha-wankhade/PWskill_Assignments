**Q1. What is the difference between a function and a method in Python? Explain with an example.**
  - In Python, both functions and methods are blocks of reusable code that perform a specific task. They help reduce code repetition and make programs easier to understand and maintain. Although they appear similar, they are used differently.
  - Function :
      - A function is an independent block of code that performs a particular task. It is defined using the def keyword and can be called directly by its name. A function is not associated with any object or class.

        
          Example of a Function
          ```
              def greet():
                  print("Welcome to Python!")

              greet()

      In this example, greet() is a function because it is defined independently and called directly.

- Method :
    - A method is a function that belongs to an object or a class. Methods are called using the object name followed by the dot (.) operator. They usually perform operations on the object's data.

      Example of a Method
      ```
      name = "python"
      print(name.upper())

    Here, upper() is a method because it belongs to the string object name.

**Q2. Explain the concept of function arguments and parameters in Python.**

  - Parameter :
      - A parameter is a variable written in the function definition. It acts as a placeholder that receives a value when the function is called.

  - Argument :
      - An argument is the actual value passed to the function during the function call.

           Example

            def greet(name):
                print("Hello,", name)

            greet("xyz")

          Here,

          name is the parameter.
          "xyz" is the argument.

   - Parameters are variables defined in the function, while arguments are the actual values passed to the function. Python supports positional, keyword, default, and variable-length arguments to make functions more flexible.

**Q5. What are iterators in Python and how do they differ from iterables?**
  - Python provides a simple way to process collections of data one element at a time using iterables and iterators.

   - **Iterable :** An iterable is any object that contains multiple elements and can be traversed one by one using a loop.

        Examples : List, Tuple, String, Dictionary, Set.

          Example:

          numbers = [10, 20, 30]

          for num in numbers:
              print(num)

          Output :
          10
          20
          30
  - **Iterator** : An iterator is an object that is created from an iterable using the iter() function. It allows elements to be accessed one at a time using the next() function.

        Example:

        numbers = [10, 20, 30]
        itr = iter(numbers)

        print(next(itr))
        print(next(itr))
        print(next(itr))

        Output :
        10
        20
        30

    If next() is called after all elements have been accessed, Python raises a StopIteration exception.

  - An iterable is a collection of elements, while an iterator is an object that helps access those elements sequentially. Iterators are memory-efficient and are especially useful when working with large amounts of data.

**Q6. Explain the concept of generators in Python and how they are defined.**
  - A generator in Python is a special type of function that generates values one at a time instead of returning all values at once. It is mainly used to work with large amounts of data efficiently because it does not store all the values in memory.
  - Unlike a normal function, which uses the return statement and ends after returning a value, a generator uses the yield keyword. When the yield statement is executed, the generator returns a value and pauses its execution. The next time it is called, it resumes from where it stopped.
  - Generators are useful when processing large datasets, reading files, or generating sequences because they save memory and improve performance.
  - A generator is defined like a normal function using the def keyword, but instead of return, it uses the yield keyword.

        Example
        def numbers():
            yield 1
            yield 2
            yield 3
            yield 4

        gen = numbers()

        print(next(gen))
        print(next(gen))
        print(next(gen))
        print(next(gen))

        Output :
        1
        2
        3
        4

**Q7. What are the advantages of using generators over regular functions?**
  - Generators are more efficient than regular functions when working with large amounts of data. Instead of creating and storing all values at once, they generate values only when required.
  - Advantages of Generators
      1. Memory Efficient : Generators do not store all values in memory. They generate one value at a time, making them suitable for large datasets.

      2. Faster Execution : Since values are generated only when needed, generators can start producing output immediately without waiting for the entire sequence to be created.

      3. Lazy Evaluation : Generators use lazy evaluation, which means values are calculated only when requested.

      4. Suitable for Large Data : Generators are useful when reading large files, processing database records, or working with large sequences.

      5. Cleaner Code : Using generators often reduces the need for complex loops and temporary variables, making the code easier to read.

        Example of a Regular Function
        def numbers():
            return [1, 2, 3, 4, 5]
        print(numbers())

        Here, the complete list is created and stored in memory.

        Example of a Generator
        def numbers():
            for i in range(1, 6):
                yield i

        for num in numbers():
            print(num)

        Here, each number is generated one by one only when needed.

 - Generators are preferred over regular functions when working with large datasets because they are memory-efficient, faster, and generate values only when required.

**Q9. Explain the purpose and usage of the map() function in Python.**
  - The map() function is a built-in Python function used to apply a specific function to every element of an iterable such as a list, tuple, or set.
  - Instead of writing loops, map() performs the same operation on each element automatically and returns a map object. The result is usually converted into a list using the list() function.
  - Syntax
      map(function, iterable)

  - Where: function – The function to apply to each element.
    iterable – The collection of elements.
     Example :
        
          numbers = [1, 2, 3, 4, 5]
          square = list(map(lambda x: x * x, numbers))
          print(square)

          Output : [1, 4, 9, 16, 25]

  - Advantages of map() :
    - Reduces the need for loops.
    - Makes the code shorter and cleaner.
    - Improves readability.
    - Can process multiple elements efficiently.

**Q10. What is the difference between map(), reduce(), and filter() functions in Python?**
 - map(), filter(), and reduce() are higher-order functions in Python that are used for processing collections of data. Although they are similar, each function serves a different purpose.

      1. map() Function : The map() function applies a given function to every element in an iterable and returns the transformed values.

        Example :

        numbers = [1, 2, 3, 4]
        result = list(map(lambda x: x * 2, numbers))
        print(result)

        Output : [2, 4, 6, 8]
      2. filter() Function : The filter() function selects only those elements that satisfy a given condition.

        Example

        numbers = [1, 2, 3, 4, 5, 6]
        result = list(filter(lambda x: x % 2 == 0, numbers))
        print(result)

        Output : [2, 4, 6]
      3. reduce() Function : The reduce() function repeatedly applies a function to the elements of an iterable until a single final value is obtained. It is available in the functools module.

        Example

          from functools import reduce
          numbers = [1, 2, 3, 4]
          result = reduce(lambda x, y: x + y, numbers)
          print(result)

          Output : 10
      



````python
#Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
```
def sum_even(numbers):
    total = 0

    for num in numbers:
        if num % 2 == 0:
            total += num

    return total


numbers = [10, 15, 20, 25, 30, 35]

result = sum_even(numbers)

print("List:", numbers)
print("Sum of Even Numbers:", result)

```
#Q2. Create a Python function that accepts a string and returns the reverse of that string.
```
def reverse_string(text):
    return text[::-1]


name = "Python"

print("Original String:", name)
print("Reversed String:", reverse_string(name))
```
#Q3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
```
def square_numbers(numbers):
    squares = []

    for num in numbers:
        squares.append(num ** 2)

    return squares


numbers = [2, 4, 6, 8]

print("Original List:", numbers)
print("Squared List:", square_numbers(numbers))
```
#Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.
```
def is_prime(number):

    if number < 2:
        return False

    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False

    return True


num = int(input("Enter a number (1-200): "))

if is_prime(num):
    print(num, "is a Prime Number")
else:
    print(num, "is Not a Prime Number")
```
#Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
```
class Fibonacci:

    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):

        if self.count >= self.terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return 0

        elif self.count == 1:
            self.count += 1
            return 1

        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b


fib = Fibonacci(10)

print("Fibonacci Series:")

for num in fib:
    print(num)
```

#Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
```
def powers_of_two(exponent):

    for i in range(exponent + 1):
        yield 2 ** i


print("Powers of 2:")

for value in powers_of_two(6):
    print(value)
```
#Q7. Implement a generator function that reads a file line by line and yields each line as a string.
```
def read_file(filename):

    with open(filename, "r") as file:
        for line in file:
            yield line.strip()


for line in read_file("sample.txt"):
    print(line)
```
#Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
```
students = [
    ("Rahul", 75),
    ("Amit", 90),
    ("Neha", 82),
    ("Priya", 68)
]

students.sort(key=lambda x: x[1])

print(students)
```
#Q9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.
```
celsius = [0, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```
#Q10. Create a Python program that uses filter() to remove all the vowels from a given string.
```
text = "Programming"

result = "".join(filter(lambda x: x.lower() not in "aeiou", text))

print("Original String:", text)
print("After Removing Vowels:", result)
```
#11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
```
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

result = list(map(
    lambda item: (
        item[0],
        item[2] * item[3] if item[2] * item[3] >= 100
        else item[2] * item[3] + 10
    ),
    orders
))

print(result)
```
````
