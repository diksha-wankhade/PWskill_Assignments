**1. Research and compare different configuration management tools (Ansible,
Puppet, Chef) and write a report on their use cases and advantages?**
1. Ansible :
    - Ansible is an open-source automation tool developed by Red Hat. It is one of the easiest configuration management tools because it does not require any agent to be installed on the target machines. It communicates with servers using SSH and uses simple YAML files called Playbooks to automate tasks.
   - Advantages of Ansible
      - Easy to install and configure.
      - No agent installation is required.
      - Uses simple YAML syntax, which is easy to understand.
      - Suitable for beginners as well as experienced users.
      - Can automate software installation, server configuration, and application deployment.
   - Use Cases
      - Server configuration
      - Software installation
      - Application deployment
      - Cloud infrastructure automation
      - CI/CD pipeline automation
2. Puppet
    - Puppet is another popular configuration management tool that follows a client-server architecture. It uses its own language called Puppet DSL to define system configurations. Puppet is mainly used in large organizations where many servers need to be managed together.
    - Advantages of Puppet :
       - Automatically maintains the desired configuration of servers.
       - Suitable for large enterprise environments.
       - Provides detailed reporting and monitoring.
       - Supports centralized management of multiple servers.
       - Highly scalable.
    - Use Cases :
       - Enterprise server management
       - Configuration enforcement
       - Security compliance
       - Data center automation
       - Large-scale infrastructure management
3. Chef
    - Chef is also an open-source configuration management tool. It uses Ruby programming language to write automation scripts known as Recipes and Cookbooks. Chef provides greater flexibility and is suitable for managing complex infrastructures.
    - Advantages of Chef :
        - Highly flexible and customizable.
        - Good support for cloud environments.
        - Efficient for managing large infrastructures.
        - Supports Infrastructure as Code (IaC).
        - Reusable automation scripts.
    - Use Cases
        - Cloud infrastructure management
        - DevOps automation
        - Continuous deployment
        - Large server environments
        - Multi-cloud deployment

**Q2. Extend the CI pipeline to include continuous delivery using Jenkins or GitLab CI. Implement a Blue-Green Deployment Strategy.**
  - Continuous Integration (CI) is a software development practice where developers regularly merge their code into a shared repository. Whenever new code is committed, the CI pipeline automatically builds the project and runs tests.
  - Continuous Delivery (CD) is the next step after Continuous Integration. In Continuous Delivery, after the build and testing stages are completed successfully, the application is automatically prepared for deployment. This reduces manual effort and helps release software faster and more reliably.
  - One of the most commonly used deployment techniques in Continuous Delivery is the Blue-Green Deployment Strategy.
  - **Blue-Green Deployment Strategy** :
      - Blue-Green Deployment is a deployment technique that uses two identical environments:
      - Blue Environment – The current version of the application that users are accessing.
      - Green Environment – The new version of the application where the latest changes are deployed.

        Initially, all users access the Blue environment. The new version of the application is deployed to the Green environment without affecting users. After testing the Green environment and confirming that everything works correctly, user traffic is switched from Blue to Green. If any issue occurs, traffic can quickly be switched back to the Blue environment.

        This method ensures almost zero downtime during deployment.

  - Implementation Using GitLab CI

    - The CI/CD pipeline consists of the following stages:

       - Developer pushes code to the GitLab repository.
       - GitLab CI automatically starts the pipeline.
       - The application is built.
       - Automated tests are executed.
       - A Docker image is created.
       - The Docker image is pushed to the Container Registry.
       - The application is deployed to the Green environment.
       - Testing is performed on the Green environment.
       - If testing is successful, production traffic is redirected from the Blue environment to the Green environment.
       - The Blue environment is kept available as a backup for rollback.
  - Advantages of Blue-Green Deployment
      - Deployment can be performed without downtime.
      - Rollback is quick and easy if any issue occurs.
      - Users experience minimal service interruption.
      - The new version can be fully tested before switching traffic.
      - Reduces deployment risk and improves application availability.
  - Applications : Blue-Green Deployment is commonly used in:

      - Banking applications
      - E-commerce websites
      - Healthcare systems
      - Cloud-based applications
      - Enterprise software
      - Financial applications
