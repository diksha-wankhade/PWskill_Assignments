**1.  Explain the key features of Python that make it a popular choice for programming.**
  - Python is a high-level, interpreted, and general-purpose programming language developed by Guido van Rossum in 1991. It is one of the most popular programming languages because of its simple syntax, readability, and wide range of applications. Python is used in web development, software development, data science, artificial intelligence, machine learning, automation, and many other fields.

  - <b>Key Features of Python:</b> <br>
    <b>1. Simple and Easy to Learn :</b>Python has a simple and easy-to-understand syntax that is similar to the English language. It allows beginners to learn programming quickly and write programs with fewer lines of code.<br>
    <b>2. Interpreted Language :</b> <br>Python is an interpreted language, which means the code is executed line by line without the need for compilation. This makes debugging easier and speeds up program development.<br>
    <b>3. Platform Independent :</b>Python programs can run on different operating systems such as Windows, Linux, and macOS without changing the source code. This makes Python a portable programming language.<br>
    <b>4. Object-Oriented Programming :</b>Python supports Object-Oriented Programming (OOP), which allows developers to create classes and objects. OOP helps organize code, improve reusability, and simplify maintenance of large applications.<br>
    <b>5. Large Standard Library :</b>Python provides a rich collection of built-in libraries and modules for tasks such as file handling, mathematical calculations, networking, database connectivity, and web development. These libraries reduce development time.<br>
    <b>6. Easy Integration :</b>Python can easily integrate with other programming languages such as C, C++, and Java. It also supports connectivity with various databases like MySQL, Oracle, PostgreSQL, and SQLite.    <br>

**3.  Compare and contrast mutable and immutable objects in Python with examples.**
  - In Python, every value is stored as an object. These objects are classified into two categories: mutable and immutable.
  - The main difference between them is whether their contents can be changed after they are created. Understanding mutable and immutable objects is important because it helps programmers write efficient programs and avoid unexpected changes in data.

  - <u>Mutable Objects :</u> Mutable objects are those whose values or contents can be modified after they are created without creating a new object. When changes are made, the same object is updated in memory.<br>
     - Common examples of mutable objects are: List, Dictionary, Set, Bytearray
  - <u>Immutable Objects :</u>Immutable objects are those whose values cannot be changed after creation. If we try to modify an immutable object, Python creates a new object instead of changing the existing one.
    - Common examples of immutable objects are: Integer, Float, String, Tuple, Boolean
  - Mutable and immutable objects are important concepts in Python. Mutable objects are useful when data needs to be updated frequently, while immutable objects provide data safety and consistency. Choosing the appropriate type depends on the application's requirements.

 **4.  Discuss the different types of operators in Python and provide examples of how they are used.**
  - Operators are special symbols used to perform operations on variables and values. They help in performing arithmetic calculations, comparisons, logical decisions, assignments, and other operations. Python provides several types of operators that simplify programming and improve code readability.

  - 1. Arithmetic Operators : Arithmetic operators perform mathematical operations.arithmatic Operators are +, - , * , /, %	, //, **
    - Example :   
    ```python
      a = 15 , b = 4
      print(a + b)
      print(a % b)
      print(a ** 2)
    ```
   - 2. Comparison (Relational) Operators : These operators compare two values and return either True or False. Operators: ==, !=, <, =, <=
     - Example :
       ```
        x = 20
        y = 10
        print(x > y)
        print(x == y)
       ```
   - 3. Logical Operators : Logical operators combine multiple conditions.
Operators: and, or, not
      - Example
        ```
         age = 20
         citizen = True
         print(age >= 18 and citizen)
        ```  
 - 4. Assignment Operators : These operators assign values to variables.
Operators: =, +=, -=, *=, /=, %=
   - Example :
      ```
        x = 10
        x += 5
        print(x)
      ```
 - 5. Membership Operators : These operators check whether a value exists in a sequence. Operators: in, not in
   - Example :
      ```
        fruits = ["Apple", "Banana", "Mango"]
        print("Banana" in fruits)
      ```
 - 6. Identity Operators : Identity operators check whether two variables refer to the same object. Operators: is, is not
   - Example :
      ```
      a = [1,2]
      b = a
      print(a is b)
      ```

**5.  Explain the concept of type casting in Python with examples.**
  - Type casting is the process of converting one data type into another. It is useful when a program needs to perform operations on different types of data. Python supports both automatic and manual type conversion.
  - Types of Type Casting
    1. Implicit Type Casting : Implicit type casting is performed automatically by Python whenever possible.
     - Example :
       ```
          a = 10
          b = 5.5

          result = a + b
          print(result)

          Output : 15.5

    - Python automatically converts the integer value into a floating-point number before performing the addition.

    2. Explicit Type Casting : Explicit type casting is performed by the programmer using built-in functions. Common functions: int(), float(), str(), list(), tuple()
     - Example :
       ```
        num = "100"

        value = int(num)

        print(value + 50)

        Output : 150
       ```
 - Advantages of Type Casting :
   - Converts incompatible data into compatible types.
   - Makes mathematical calculations easier.
   - Improves program flexibility.
   - Helps process user input correctly.


**6.  How do conditional statements work in Python? Illustrate with examples.**
   - Conditional statements are used to make decisions in a Python program. They allow the program to execute different blocks of code depending on whether a condition is True or False. Conditional statements make programs interactive and intelligent.

   - Python provides the following conditional statements:
     - if Statement : The if statement executes a block of code only when the specified condition is true.
        - Example
           ```
             age = 20
             if age >= 18:
               print("Eligible to Vote");

           ```
           
     - if-else Statement : The if-else statement executes one block if the condition is true and another block if it is false.
        - Example
            ```
              marks = 35
              if marks >= 40:
                print("Pass")
              else:
                print("Fail")
     - if-elif-else Statement : This statement is used when multiple conditions need to be checked.
        - Example :
            ```
              marks = 82
              if marks >= 90:
                 print("Grade A+")
              elif marks >= 75:
                 print("Grade A")
              elif marks >= 60:
                 print("Grade B")
              else:
                 print("Grade C")
   - Conditional statements help Python programs make decisions based on different situations. They improve the flexibility of programs and are widely used in almost every real-world application.

**7.  Describe the different types of loops in Python and their use cases with examples.**
  - Loops are used to execute a block of code repeatedly until a specified condition is satisfied. They reduce code duplication, save development time, and make programs more efficient. Python mainly provides two types of loops: for loop and while loop.
    1. for Loop : A for loop is used when the number of iterations is known. It is commonly used to iterate through sequences such as lists, tuples, strings, dictionaries, and ranges.
      - Example
          ```
            for i in range(1,6):
                print(i)

            Output:
                1
                2
                3
                4
                5

    - Use Cases :
        - Printing numbers.
        - Traversing a list or string.
        - Reading data from files.
        - Repeating a task a fixed number of times.
    2. while Loop : A while loop is used when the number of iterations is not known in advance. It continues executing until the specified condition becomes false.
        - Example
            ```
              i = 1
              while i <= 5:
                print(i)
                i += 1
            Output :
                  1
                  2
                  3
                  4
                  5
    

    - Use Cases :
        - Menu-driven applications.
        - Taking user input until valid data is entered.
        - Processing data until a condition changes.
        - Waiting for an event to occur.
