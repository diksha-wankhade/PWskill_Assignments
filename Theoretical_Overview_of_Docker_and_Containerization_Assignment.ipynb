````python
# commands are in bash
#Step 1: Install Docker
```
sudo apt update
sudo apt install docker.io -y
```
#Start and enable the Docker service.
```
sudo systemctl start docker
sudo systemctl enable docker
```
#Verify the Docker installation.
```
docker --version
```
#To ensure Docker is working correctly, run the following command:
```
docker run hello-world
```
#step 2: Create a new project folder.
```
mkdir docker-demo
cd docker-demo
```
#Step 3: Create a Python Application
```
#Create a file named app.py.
```
nano app.py
```
#Adding the following Python code in app.py:
```
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return "Docker Container is Running Successfully!"

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
#Step 4: Create requirements.txt

#Create a dependency file.
```
nano requirements.txt
```
#Add the following content:
```
Flask
```
#Step 5: Create a Dockerfile

#Create a file named Dockerfile.
```
nano Dockerfile
```
#Add the following content:
```
FROM python:3.11-slim

WORKDIR /app

COPY . .

RUN pip install --no-cache-dir -r requirements.txt

EXPOSE 5000

CMD ["python", "app.py"]
```
#Step 6: Build the Docker Image

#Build the Docker image using the Dockerfile.
```
docker build -t flask-demo .
```
#Step 7: Verify Docker Image

#List all Docker images.
```
docker images
```
#Step 8: Run the Docker Container

#Run the container.
```
docker run -d -p 5000:5000 --name flask-container flask-demo
```
#Check whether the container is running.
```
docker ps
```

````
